```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
import time

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

#Developing a content-based recommender system using the genres and/or descriptions.
```

```python
df = pd.read_csv('titles.csv')
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['description'].fillna('Sin descripción', inplace=True)
df['age_certification'].fillna('Desconocido', inplace=True)
df['seasons'].fillna(0, inplace=True)
df.drop(['imdb_id', 'imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score'], axis=1, inplace=True)
```

```python
df.isna().sum()
```

```python
df['genres'] = df['genres'].str.replace("[", "").str.replace("]", "").str.replace("'", "").str.split(", ")
```

```python
unique_genres = set(x for _list in df['genres'] for x in _list)
for genre in unique_genres:
    df[genre] = df['genres'].apply(lambda x: 1 if genre in x else 0)
    
```

```python
df.info()
```

```python
df
```

```python
df['description'].head()
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
df['description'] = df['description'].fillna('')

tfidf_matrix = tfidf.fit_transform(df['description'])

tfidf_matrix.shape
```

```python
#utilizaremos la similitud del coseno

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
#Constrir un mapa inverso de indices y titulos de los animes
indices = pd.Series(df.index, index=df['title']).drop_duplicates()
```

```python
def get_recomendations(title, cosine_sim = cosine_sim):
    idx = indices[title]
    
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    sim_scores = sim_scores[1:11]
    
    anime_indices = [i[0] for i in sim_scores]
    
    return df['title'].iloc[anime_indices]
```

```python
get_recomendations('Gunbuster')
```
